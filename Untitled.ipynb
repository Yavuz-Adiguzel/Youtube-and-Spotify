<h1 style:"15">YOUTUBE VE SPOTİFY EDA ANALİZ</h1>

***Ben bu projede Youtube ve Spotify üzerindeki verileri, aykırı değerleri, veriler arasındaki ilişkiyi EDA ile göstermeye çalıştım. Aynı zamanda verileri görselleştirmeye çalıştım.***


```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(color_codes=True)
import datetime as dt
import plotly.express as px
import warnings 
warnings.simplefilter(action='ignore')
```

```python
data=pd.read_csv('Spotify_Youtube.csv')
data.head()
```

```python
data.info()
```

```python
#Toplam boş verileri kontrol etmek için
data.isnull().sum()
```

```python
#Sütunlara bakmak için;
data.columns
```

```python
#Sütunlardaki benzersiz verilere bakmak için;
data.nunique()
```

```python
#İşimize yaramayan sütunları kaldırmak için;
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data.drop(['Uri'],axis=1,inplace=True)
data.drop(['Url_youtube'],axis=1,inplace=True)
data.drop(['Title'],axis=1,inplace=True)
data.drop(['Description'],axis=1,inplace=True)
```

```python
#Bazı sütunların isimlerini değiştirmek için;
data.rename(columns={'official_video':'Official Video'},inplace=True)
data.rename(columns={'Album_type':'Album Type'},inplace=True)
data.rename(columns={'Duration_ms':'Duration'},inplace=True)
```

```python
#Sütunlarda tekrar eden veri var mı diye kontrol etmek için;
data.duplicated().sum()
```

```python
#eksik verileri silmek için;
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
#Veritiplerini görmek için;
data.dtypes
```

```python
#Verinin daha iyi okunabilir ve anlaşılır olması için veri tipini Stringe çevrilmesi;
data['Licensed']=data['Licensed'].astype('str')
data['Official Video']=data['Official Video'].astype('str')
```

```python
#Şarkıların süresini dakika cinsinden gösteren yeni sütun oluşturmak için;
data['Duration Minutes']=data['Duration']/60000
```

```python
data.head()
```

```python
#Değerleri daha kolay okuyabilmek için datanın düzenlenmesi;
data['Official Video']=data['Official Video'].str.replace('1','Yes')
data['Official Video']=data['Official Video'].str.replace('0','No')
data['Licensed']=data['Licensed'].str.replace('1','Yes')
data['Licensed']=data['Licensed'].str.replace('0','No')
```

```python
data.head()
```

```python
data.describe()
```

```python
#Aykırı değer kontrolü
#Sadece sayısal sütunları seçmek için;
numeric_data=data.select_dtypes(include=['float64','int64'])

#Z skoru hesaplama;
z_scores=np.abs(stats.zscore(numeric_data))

#Z skoru 3 ten büyük olanlar aykırı olarak kabul edilir. Bunu görmek için;
outliers=np.where(z_scores>3)

#Aykırı değerlerin indekslerini ve sütunlarını görmek için;
outlier_indices=outliers[0]
outlier_columns=outliers[1]

print(f'Aykırı değerlerin bulunduğu satır sayısı:{len(set(outlier_indices))}')
print(f'Aykırı değerlerin bulunduğu sütunlar:{numeric_data.columns[outlier_columns]}')
```

```python
outlier_columns=['Likes','Stream','Views','Speechiness','Loudness','Liveness','Instrumentalness']

#Her bir sütun için boxplot oluşturmak için;
plt.figure(figsize=(15,10))

for i, column in enumerate(outlier_columns,1):
    plt.subplot(3,3,i)
    sns.boxplot(x=data[column],color='lightblue')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()#Bu satır, tüm grafiklerin birbirine çok yakın olmadan, daha düzenli bir şekilde yerleşmesini sağlar.

plt.show()
```

```python
columns_to_check = ['Likes', 'Stream', 'Views']

#Aykırı değerleri Z skor ile bulmak için;
z_scores=np.abs(stats.zscore(data[columns_to_check]))

#Z skoru 3 ten büyük verileri bulmak için;
outliers=(z_scores>3)

#Aykırı değerlerin toplamını bulmak için;
total_outliers=np.sum(outliers)

#Her bir sütundaki aykırı değerleri toplamak için;
outliers_per_column=np.sum(outliers , axis=0)

print(f'Aykırı değerlerin toplamı:{total_outliers}')
print(f'Sütun bazında aykırı değerler:\n{outliers_per_column}')
```

```python
columns_to_check = ['Likes', 'Stream', 'Views']

#Her sütun için z skorlarını bulma
data['Likes_zscore']=np.abs(stats.zscore(data['Likes']))
data['Stream_zscore']=np.abs(stats.zscore(data['Stream']))
data['Views_zscore']=np.abs(stats.zscore(data['Views']))

data_outliers=data[(data['Likes_zscore']>3) | (data['Stream_zscore']>3) | (data['Views_zscore']>3)]

#Aykırı değerlerini z skoruna göre sıralama;
data_outliers_sorted = data_outliers.sort_values(by=['Likes_zscore','Stream_zscore','Views_zscore'],ascending=False)

#En uçtaki 10 aykırı değeri görüntülemek için;
data_outliers_sorted[['Track', 'Artist', 'Likes', 'Stream', 'Views', 'Likes_zscore', 'Stream_zscore', 'Views_zscore']].head(10)
```

<h1 style:'15'>En çok izlenen şarkılara göre sıralama</h1>

```python
#Sanatçıları views bazında gruplandırma ve toplam izlenme sayısına göre sıralama
top_10_artists_view=data.groupby('Artist')['Views'].sum().sort_values(ascending=False).head(10)

#Barplot oluşturma
plt.figure(figsize=(10,6))
sns.barplot(x=top_10_artists_view.values, y=top_10_artists_view.index, palette='viridis')

plt.title('Top 10 Most Viewed Artists', fontsize=16)
plt.xlabel('Total Views', fontsize=12)
plt.ylabel('Artist', fontsize=12)

plt.tight_layout()
plt.show()
```

***Sanatçının tüm şarkılarının izlenme sayıları toplandığında sıralama bu şekilde oluyor, ancak bu değerlendirme adil değil çünkü sanatçıların farklı sayıda şarkıları olabilir. Bu nedenle, sanatçının en popüler şarkısını bularak izlenme sayılarına göre sıralama yapmak daha mantıklı olacaktır.***

```python
#Her sanatçının en popüler şarkısını bulma
top_songs_per_artist=data.loc[data.groupby('Artist')['Views'].idxmax()]

# En popüler şarkıya göre en yüksek izlenme sayısına sahip 10 sanatçıyı bulma
top_10_artist_views1=top_songs_per_artist[['Artist','Views']].sort_values(by='Views',ascending=False).head(10)

#Barplot oluşturma
plt.figure(figsize=(10,6))
sns.barplot(x=top_10_artist_views1['Views'],y=top_10_artist_views1['Artist'],palette='viridis')

plt.title('Top 10 Most Viewed Artists (Based on Their Most Popular Song)', fontsize=16)
plt.xlabel('Views (Most Popular Song)', fontsize=12)
plt.ylabel('Artist', fontsize=12)

plt.tight_layout()
plt.show()
```

<h1 style:'15'>En Çok Etkileşim Alan Youtube Kanal Analizi</h1>

```python
#Youtube verilerini kanal bazında gruplayarak toplam beğeni,izlenme ve yorum sayısını hesaplama
channel_groupped=data.groupby('Channel')[['Likes','Views','Comments']].sum().reset_index()

#Likes,Views ve Comments sütunlarına göre en çok etkileşim olan 10 kanalı bulma
top_channels_likes = channel_groupped.sort_values(by='Likes', ascending=False).head(10)
top_channels_views = channel_groupped.sort_values(by='Views', ascending=False).head(10)
top_channels_comments = channel_groupped.sort_values(by='Comments', ascending=False).head(10)

plt.figure(figsize=(10,15))

#Likes barplot
plt.subplot(3, 1, 1)
sns.barplot(x=top_channels_likes['Likes'], y=top_channels_likes['Channel'], palette='magma')
plt.title('Top 10 Channels by Likes')
plt.xlabel('Likes')
plt.ylabel('Channel')

#Views barplot
plt.subplot(3,1,2)
sns.barplot(x=top_channels_views['Views'],y=top_channels_views['Channel'],palette='viridis')
plt.title('Top 10 Channels by Views')
plt.xlabel('Views')
plt.ylabel('Channel')

#Comment barplot
plt.subplot(3,1,3)
sns.barplot(x=top_channels_comments['Comments'],y=top_channels_comments['Channel'],palette='plasma')
plt.title('Top 10 Channels by Comment')
plt.xlabel('Comments')
plt.ylabel('Channel')

plt.tight_layout()
plt.show()
```
